```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.hist(df["Rating"].dropna(),bins=20)
plt.title("App Ratings Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

Most apps in this datset are highly rated with the typical app scoring around a 4.3 to 4.5

```python
df["Category"].value_counts().head(10).plot(kind="bar")
plt.xlabel("Categories")
plt.ylabel("Number of Apps")
plt.title("Top 10 App Categories")
plt.show()
```

This bar chart titled"Top 10 App Categories",provides a visual representation of the most populated categories in the data set based on the total number of apps.Family having the highest count(nearly 1750) and the other categories ranging down lowest (Medical and Photography )

```python
plt.scatter(df["Reviews"],df["Rating"])
plt.xlabel("Reviews")
plt.ylabel("Rating")
plt.title("Reviews vs Rating")                  
plt.show()
```

The graph suggests that having more reviews does not guarantee a higher rating 

```python
df["Type"].value_counts().plot(kind="pie",autopct="%1.2f%%")
plt.title("Free vs Paid Apps")
plt.show()
```

```python
sns.boxplot(x="Type",y="Rating",data=df)
plt.title("Rating comparison by app type")
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.title("correlation heatmap")
plt.show()
```

```python
plt.scatter(df["Reviews"],df["Installs"])
plt.xlabel("Reviews")
plt.ylabel("Installs")
plt.title("Reviews vs installs")
plt.show()
```

```python
review_install=df.groupby("Category")[["Reviews","Installs"]].mean()
```

```python
review_install
```

```python
plt.scatter(review_install["Reviews"],review_install["Installs"])
plt.title("Reviews vs installs")
plt.xlabel("Reviews")
plt.ylabel("Installs")
plt.show()
```

```python
sns.lineplot(data=review_install,x="Reviews",y="Installs")
```

```python
reviews_rating=df.groupby("Category")[["Reviews","Rating"]].mean()
```

```python
reviews_rating.sort_values(by="Reviews",ascending=False)
```

```python
reviews_rating
```

```python
reviews_rating.head(10)["Rating"].plot(kind="bar",figsize=(10,5))
plt.xlabel("Category")
plt.ylabel("average of rating")
plt.title("Top 10 categories average installs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
reviews_rating.head(10)["Installs"].plot(kind="bar",figsize=(10,5))
plt.xlabel("Category")
plt.ylabel("average of installs")
plt.title("Top 10 categories average installs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
